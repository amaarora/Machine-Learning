```python
#Steps before plotting to make more informative charts

accidents['Weather_Conditions'].replace(-1, 9, inplace = True)

weather_values = {
                    1:"Fine no high winds",
                    2:"Raining no high winds",
                    3:"Snowing no high winds",
                    4:"Fine + high winds",
                    5:"Raining + high winds",
                    6:"Snowing + high winds",
                    7:"Fog or mist",
                    8:"Other",
                    9:"Unknown"
                 }

def alphabetical_order_col(df,col_name):
    ordered_list = df[col_name].sort_values().unique()
    cleaned_list = [x for x in ordered_list if str(x) != 'nan']
    return cleaned_list
```

**Number of Casualties by Weather Conditions:**

```python
accidents['Weather'] = accidents['Weather_Conditions'].map(weather_values)

fig,ax = plt.subplots(figsize = (10,6))

a = sns.countplot(x='Weather', 
                  data = accidents, 
                  order = alphabetical_order_col(accidents,'Weather')); #Function call to get labels

ax.set(ylabel = 'Number_of_Casualties');

#Rotate xlabel
for item in a.get_xticklabels():
    item.set_rotation(-90)
```

The most common weather type is 'Fine no high winds' and as a result the most amount of casualties are also caused in this weather. 

```python
mask = (accidents['Weather'] != 'Unknown')

accidents[mask].groupby('Weather')['Number_of_Casualties'].mean().plot(kind = 'barh', 
                                                                       figsize = (10,6));

plt.xlabel('Average number of Casualties for each accident that occured');
```

Even though 'Snowing + high winds' is not the most common weather type, I would be the most careful if for some reason I have to drive under this weather condition as it is the most risky weather type.

```python
#read and merge new data
model_data_path = '/Users/user/Desktop/Folders/Data_Scientist/Github Desktop_Repo_PTDS/intro-ds/projects/Project 2/data/uk-2016-road-safety-data/MakeModel2016.csv'
model = pd.read_csv(model_data_path)

#merge with data to include model_data
master_data = pd.DataFrame.merge(master_data, model, how = 'inner')

#regenerating mask for car,bike,bicicyle using vehicle_type col
mask_car = (master_data['Vehicle_Type'].isin((8,9)))

mask_bike = (master_data['Vehicle_Type'].isin((2,3,4,5,23,97)))

mask_bicycle = (master_data['Vehicle_Type'] == 1)
```

```python
#calculating proportion
list_of_models=[]
x = master_data[mask_car]['make'].value_counts()[:5]
y = master_data[mask_bike]['make'].value_counts()[:5]
list_of_models.append(x.index.values)
list_of_models.append(y.index.values)

array_of_models = []
for x in list_of_models:
    for y in x:
        array_of_models.append(y)

total_accidents = master_data.shape[0]
count=0
for x in master_data['make']:
    if x in array_of_models:
        count+=1
print('Proportion of Accidents for these 10 vehicle makers:',(count/total_accidents)*100)
```
